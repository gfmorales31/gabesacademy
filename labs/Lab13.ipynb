## Task 2
1. Define a class named `Vehicle` that would be appropriate for representing a vehicle. 
    1. Include attributes named `make` (a string), `model` (a string), and `body_type` (a string).
    1. Include a `__str__()` method to display information about a `Vehicle`
1. Define an equals method `__eq__()` to compare 2 `Vehicle`s for equality. The method shoud return `True` if they are the same vehicle and `False` if they are different vehicles.
1. Write a program that initializes 2 `Vehicle` objects and compares them for equality. Compare for equality using two approaches:
    1. `print(vehicle1.__eq__(vehicle2))` to test for equality between two `Vehicle` objects, `vehicle1` and `vehicle2`, by explicitly calling the `__eq__()` special method.
    1. `print(vehicle1 == vehicle2)` to test for equality between two `Vehicle` objects, `vehicle1` and `vehicle2`, by implicitly calling the `__eq__()`. This approach is called *operator overloading*. If an `__eq__()` method is defined for `Vehicle`, then a comparison between `Vehicle` objects with the `==` operator invokes the `__eq__()` method.
    
Example output:
```
Vehicle 1:
Make: Toyota
Model: Tacoma
Body type: Truck

Vehicle 2:
Make: Toyota
Model: 4Runner
Body type: SUV

Vehicle 3:
Make: Toyota
Model: 4Runner
Body type: SUV

v1.__eq__(v2): False
v2 == v3: True
```
