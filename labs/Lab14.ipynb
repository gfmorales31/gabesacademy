## Task 3
Write a program to sort a list without using the `sort()` list method! Here's how to do it:
1. Start with the list to sort called `original_list`
1. Create an empty list called `sorted_list`. `sorted_list` will store `original_list` in sorted order
1. While `original_list` is not empty:
    1. Find the smallest value (and the index of the smallest value) in `original_list`
    1. Use the list method `append()` to append a value to `sorted_list`, e.g. `sorted_list.append(smallest_value)`
    1. Remove the smallest value from `original_list` using the value's index, e.g. `original_list.pop(smallest_index)`
1. Display `sorted_list`

Example output:
```
original_list:  [11, 1, 12, -3, 4, -1, 10, -5, 2]
sorted_list:  [-5, -3, -1, 1, 2, 4, 10, 11, 12]
```

Congrats! You just implemented the [Selection Sort algorithm](https://en.wikipedia.org/wiki/Selection_sort)!

Note: Do not use the reserved keyword `in` in your solution. Also, do not use any string methods! You don't need them

Here is an outline of the algorithm to do this:
1. Open [words.txt](http://thinkpython2.com/code/words.txt) for reading
1. Initialize a dictionary: `substring_dict = {"are": [], "ing": [], "the": []}`
1. For each `word` in [words.txt](http://thinkpython2.com/code/words.txt):
    1. For `substring` in `substring_dict`:
        1. If `is_substring(word, substring)` returns `True` (this is a function you wrote for a previous lab task :))
           1. Append `word` to `substring_dict[substring]`
